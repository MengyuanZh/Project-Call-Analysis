```python
import csv
import pandas as pd
```

```python
df = pd.read_csv("311_Service_Requests_2020.csv")
```

```python
# Extracting the rows that are related to our chosen zip code 10027
df_10027 = df[df["Incident Zip"] == 10027]
df_10027.head(5)
```

```python
# count the number of compaint types in the "Complaint Type" column
causes = df_10027.groupby("Complaint Type")["Complaint Type"].count().sort_values(ascending=False)
```

```python
# Find out the top 10 causes
top10 = pd.Series(causes.head(10))
top10
```
